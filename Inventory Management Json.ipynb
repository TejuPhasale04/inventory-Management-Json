```python
#importing Libraries
import json
import time

#initilizing Varible
sales=""

#Loading Inventory Record from JSON to String
fd=open("Record.json",'r')
js=fd.read()
fd.close()

#Converting String Record to JSON
record=json.loads(js)  #loads used to store and exchange data


#Displaying List of item
print("--------------------MENU-------------------------------")
for key in record.keys():
    print(key,":  ",  record[key]["name"]," |",record[key]["Qn"],"|",record[key]["price"])

#Taking Iput From User
print("------------------------------------------------------")
ui_name=   str(input("Enter Your Name           :"))
ui_email=  str(input("Enter Your Email          :"))
ui_phoneNo=str(input("Enter Phone No            :"))
ui_pr=     str(input("Enter Product Id          :"))
ui_Qn=     int(input("Enter Product Quantity    :"))

#If we're having equal or more quantity then user wants
if record[ui_pr]["Qn"] >= ui_Qn :
    
    #Printing Bill
    print("-----------------------Bill---------------------------")
    print("Name                   : ",record[ui_pr]["name"])
    print("Price                  : ",record[ui_pr]["price"])
    print("Quantity               : ",ui_Qn)
    print("------------------------------------------------------")
    print("Total Amount           : ",record[ui_pr]["price"] * ui_Qn)
    print("Discount               : ","10%"")

    
    #Updating Record
     record[ui_pr]["Qn"] =record[ui_pr]["Qn"] - ui_Qn
    
    #Generating Sales Structure CSV
    sales=1,ui_name,ui_email,ui_phoneNo,record[ui_pr]["name"],ui_Qn,record[ui_pr]["price"] * ui_Qn,time.ctime()+"\n"
    sales_str = ','.join(map(str, sales))
else:
    #If we're having no enough quantity
    print("Sorry,We're not having enough Quantity in our Inventory..")
    print("We're only having ",record[ui_pr]['Qn'],'quantity.')
    print("--------------------------------------------------")
    
    #If user wants to purches whole quantity for that product
    ch=str(input("Press Yes Or No to Purches:"))

    if(ch=="Yes" or ch=="yes"):
        #Displat Bill
        print("-----------------------Bill---------------------------")
        print("Name          :",record[ui_pr]["name"])
        print("Price         :",record[ui_pr]["price"])
        print("Quantity      :",record[ui_pr]["Qn"])
        record[ui_pr]["Qn"] =0
        print("------------------------------------------------------")
        print("Total:",record[ui_pr]["price"] * record[ui_pr]["Qn"])
        
        #Generating Sales Structure CSV
        sales=1,ui_name,ui_email,ui_phoneNo,record[ui_pr]["name"],record[ui_pr]["Qn"],record[ui_pr]["price"] * ui_Qn,time.ctime()+"\n"
        sales_str = ','.join(map(str, sales))
    else:
        print("Thank You....")

#Converting Dictionary to String
js=json.dumps(record)  #Allow to store data directly into file

#Saving Record to file
fd=open("Record.json",'w')
fd.write(js)
fd.close()

#Saving records into Sales.cvs
fd=open("Sales.csv",'a')
fd.write(sales_str)
fd.close()


print("-------------------------------------------------------------------------------------")
print("-------------------Thank You For Ordering,Inventory Updated--------------------------")
print("-------------------------------------------------------------------------------------")


```
